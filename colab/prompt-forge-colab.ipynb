## 2. (Opcional) Configurar API keys para ejecutar prompts

Si quieres que Colab **ejecute** el prompt (no solo lo compile), agrega al menos una API key.
Obtén una gratis de:
- **Gemini** (recomendado): https://aistudio.google.com
- **Groq** (más rápido): https://console.groq.com

Colab tiene un gestor seguro de secrets (🔑 icono en la barra lateral). Nombra el secret `GEMINI_API_KEY` o `GROQ_API_KEY` y actívalo para este notebook.

```python
import os

# Leer secrets de Colab si están disponibles
try:
    from google.colab import userdata
    for key in ['GEMINI_API_KEY', 'GROQ_API_KEY', 'CEREBRAS_API_KEY']:
        try:
            os.environ[key] = userdata.get(key)
            print(f'✓ {key} loaded')
        except Exception:
            pass
except ImportError:
    pass

# Verifica qué tienes disponible
!python forge.py runners
```

## 3. Compilar un prompt

```python
# @title Escribe tu petición { display-mode: 'form' }

raw_request = 'ayudame a escribir un correo a mi jefe pidiendo vacaciones' # @param {type:'string'}
execute = True  # @param {type:'boolean'}
runner = 'auto' # @param ['auto', 'gemini', 'groq', 'cerebras']

if execute:
    !python forge.py run "{raw_request}" --runner {runner} --format md
else:
    !python forge.py compile "{raw_request}" --format md
```

## 4. Descargar los archivos generados

Los archivos se guardan en `/root/prompt-forge-outputs/`. Usa esta celda para descargarlos.

```python
import os, shutil
from pathlib import Path

out_dir = Path.home() / 'prompt-forge-outputs'
files = sorted(out_dir.glob('*')) if out_dir.exists() else []
print(f'{len(files)} file(s):')
for f in files[-5:]:
    print(f'  {f.name}')

# Zip them all for easy download
if files:
    zip_path = '/content/prompt-forge-outputs.zip'
    shutil.make_archive('/content/prompt-forge-outputs', 'zip', out_dir)
    print(f'\n✓ Zipped to {zip_path}')
    try:
        from google.colab import files as colab_files
        colab_files.download(zip_path)
    except ImportError:
        print('Manual download: right-click the file in the left panel → Download')
```
